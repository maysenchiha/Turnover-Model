<h3 align="left"><font color=#4D1C87>Features Description:</font></h3>
To predict employee turnover using our dataset, hereâ€™s how each variable might influence this prediction:

__Satisfaction Level:__ A low level of employee satisfaction is often a strong indicator of the intention to leave the company. Dissatisfied employees are more likely to seek new opportunities elsewhere.

__Last Evaluation:__ A low or high evaluation score could influence the decision to leave. A low score might indicate poor performance, while a high score without recognition (like a promotion) could lead to frustration.

__Number of Projects:__ Employees overloaded with projects may feel exhausted, increasing their likelihood of leaving. However, a low number of projects could also signal a lack of challenge, pushing the employee to look for a more stimulating job.

__Average Monthly Hours:__ A high number of monthly hours can be a sign of overwork, leading to an early departure. Conversely, a low number of hours might indicate employee disengagement.

__Time Spent at Company:__ Employees who have spent several years in the company might be less inclined to leave due to job security and accumulated benefits. However, the lack of career progression over a long period could push them to leave.

__Work Accident:__ Employees who have experienced a work accident might have negative feelings toward the company, especially if the accident was poorly managed, increasing their likelihood of leaving.

__Promotion Last 3 Years:__ The absence of promotion can lead to a feeling of stagnation, encouraging employees to seek opportunities elsewhere. Those who have been recently promoted are generally more engaged and less likely to leave.

__Department:__ Some departments may have higher turnover rates due to the nature of the work or department culture.

__Salary:__ A low salary may push employees to leave for better-paying opportunities, while a high salary may encourage them to stay.

__Sex:__ Although gender itself should not influence the decision to leave, it may interact with other factors such as satisfaction or promotion opportunities.

__Marital Status:__ Married employees may be more stable in their jobs due to family responsibilities, while single or divorced employees might be more mobile.

__Distance from Home:__ A long commute can contribute to fatigue and dissatisfaction, increasing the likelihood of departure.

__Age:__ Younger employees may be more inclined to change jobs to advance their careers, while older employees might prefer stability.

__Work Mode:__ Employees working in hybrid or remote modes may be more satisfied with their work-life balance, reducing their likelihood of leaving.

__Job Position:__ Employees in high-level positions (such as Manager or Director) may have increased loyalty due to their status, although high pressure and expectations could also lead them to leave.

__Contract Type:__ Employees on temporary contracts (CDD) may be more inclined to leave, seeking job security through a permanent contract (CDI).

__Expected Promotion:__ If an employee expects a promotion but does not receive it, this could lead to disappointment and encourage them to leave the company.

__Salary Increase Last Year:__ A low or no salary adjustment can create a sense of unfairness or undervaluation, pushing the employee to leave.

__Leave Days Taken:__ A high number of leave days may indicate a need for time away from work, possibly due to stress or dissatisfaction, which could lead to a departure.

__Absenteeism:__ A high rate of absenteeism could be a sign of disengagement, which might precede a resignation.

These variables can have direct or indirect effects on an employee's decision to leave the company. By combining them, we can get a clearer picture of risk factors and thus better predict potential employee turnover.

```python
import pandas as pd
import numpy as np
from scipy.stats import beta
 
np.random.seed(42)
 
num_observations = 14000
 
proportion_not_satisfied = 0.35
proportion_satisfied = 0.45
proportion_very_satisfied = 0.2
 
num_not_satisfied = int(num_observations * proportion_not_satisfied)
num_satisfied = int(num_observations * proportion_satisfied)
num_very_satisfied = num_observations - num_not_satisfied - num_satisfied
 
not_satisfied = np.random.uniform(0, 0.3, num_not_satisfied)
satisfied = np.random.uniform(0.3, 0.8, num_satisfied)
very_satisfied = np.random.uniform(0.8, 1, num_very_satisfied)
satisfaction_level = np.concatenate([not_satisfied, satisfied, very_satisfied])
np.random.shuffle(satisfaction_level) 
last_evaluation = np.random.normal(0.6, 0.2, num_observations)
last_evaluation = np.clip(last_evaluation, 0, 1)  # Ensure values are within the [0, 1] range
number_project = np.random.randint(2, 7, size=num_observations)
average_monthly_hours = satisfaction_level * 200 + np.random.normal(50, 10, num_observations)
average_monthly_hours = np.clip(average_monthly_hours, 100, 300)
time_spend_company = np.random.randint(1, 15, size=num_observations)
Work_accident = np.random.choice([0, 1], size=num_observations, p=[0.9, 0.1])
promotion_last_3_years = np.random.choice([0, 1], size=num_observations, p=[0.95, 0.05])
absent_per_month = np.random.randint(0, 10, size=num_observations)
department = np.random.choice(['sales', 'DSI', 'hr', 'technical', 'support', 'accounting', 'marketing'], size=num_observations)
salary = np.random.choice(['low', 'medium', 'high'], size=num_observations, p=[0.3, 0.5, 0.2])
sex = np.random.choice(['Male', 'Female'], size=num_observations, p=[0.6, 0.4])
marital_status = np.random.choice(['married', 'single', 'divorced'], size=num_observations)
distance_from_home = np.random.randint(1, 50, size=num_observations)
age = np.random.randint(22, 65, size=num_observations)
job_positions = np.random.choice(['Junior', 'Senior', 'Manager', 'Director'], size=num_observations, p=[0.5, 0.4, 0.07, 0.03])
work_mode = pd.Series(department).apply(lambda x: 'office' if x == 'DSI' else np.random.choice(['home', 'office', 'hybride'], p=[0.3, 0.3, 0.4]))
contract_types = np.random.choice(['CDI', 'CDD'], size=num_observations, p=[0.7, 0.3])
expected_promotion = np.random.choice([0, 1], size=num_observations, p=[0.9, 0.1])
 
# Create the DataFrame
data = {
    'satisfaction_level': satisfaction_level,
    'last_evaluation': last_evaluation,
    'number_project': number_project,
    'average_monthly_hours': average_monthly_hours,
    'time_spend_company': time_spend_company,
    'Work_accident': Work_accident,
    'promotion_last_3_years': promotion_last_3_years,
    'department': department,
    'salary': salary,
    'sex': sex,
    'marital_status': marital_status,
    'distance_from_home': distance_from_home,
    'age': age,
    'absent_per_month': absent_per_month,
    'work_mode': work_mode,
    'job_position': job_positions,
    'contract_type': contract_types,
    'expected_promotion': expected_promotion
}
 
df = pd.DataFrame(data)
df.loc[df['satisfaction_level'] < 0.2, 'salary'] = np.nan 
 
missing_features = ['last_evaluation', 'average_monthly_hours', 'satisfaction_level']
for feature in missing_features:
    missing_indices = np.random.choice(df.index, size=int(0.05 * num_observations), replace=False)
    df.loc[missing_indices, feature] = np.nan
  
df['satisfaction_level'] += 0.05 * df['promotion_last_3_years'] + 0.1 * df['last_evaluation'] + \
                            0.05 * df['salary'].map({'low': -0.1, 'medium': 0, 'high': 0.1}) + \
                            0.02 * (df['work_mode'] == 'home').astype(int)
df['satisfaction_level'] = np.clip(df['satisfaction_level'], 0, 1)
  
last_evaluation = np.tanh((last_evaluation - 0.5) * 2) / np.tanh(1)
last_evaluation = np.clip(last_evaluation, 0, 1)  # Assurez-vous que les valeurs sont dans la plage [0, 1] 
 
df['last_evaluation'] = np.clip(df['last_evaluation'], 0, 1)
df['average_monthly_hours'] += 10 * df['number_project'] + 5 * df['marital_status'].map({'married': 1, 'single': 0, 'divorced': -1}) + \
                               0.5 * df['distance_from_home']
df['average_monthly_hours'] = np.clip(df['average_monthly_hours'], 100, 300)

outlier_indices = np.random.choice(df.index, size=10, replace=False)
df.loc[outlier_indices, 'average_monthly_hours'] = np.random.randint(400, 600, size=10)
 
df['salary'] = df.apply(lambda x: 'high' if (x['time_spend_company'] > 5 and x['job_position'] in ['Manager', 'Director'])
                                    else 'medium' if (x['time_spend_company'] >= 3 and x['time_spend_company'] <= 5 and x['job_position'] == 'Senior')
                                    else x['salary'], axis=1)
df['salary_increase_last_year'] = (
    0.1 * df['satisfaction_level'] +  # Le niveau de satisfaction affecte l'augmentation de salaire
    0.1 * df['last_evaluation'] +  # La derniÃ¨re Ã©valuation contribue aussi Ã  l'augmentation
    0.1 * (df['time_spend_company'] / 10) +  # Les annÃ©es passÃ©es dans l'entreprise sont Ã©galement un facteur
    0.1 * (df['job_position'].map({'Junior': 0.1, 'Senior': 0.2, 'Manager': 0.3, 'Director': 0.4})) +  # Les positions de travail
    np.random.normal(0.05, 0.02, num_observations)  # Ajout d'une composante alÃ©atoire pour varier
)
  
df['salary_increase_last_year'] = np.clip(df['salary_increase_last_year'], 0, 0.5)
 
df['leave_days_taken'] = np.random.randint(0, 31, size=num_observations)
 
promotion_prob = np.clip(df['last_evaluation'] * 0.5, 0, 1)  # Calculate promotion probability based on evaluation
 
promotion_prob = promotion_prob.fillna(0)  
 
df['expected_promotion'] = np.random.binomial(1, promotion_prob)
 
df['left'] = (
    (df['leave_days_taken'] > 15) * 0.2 +
    (df['satisfaction_level'] < 0.3) * 0.2 +
    (df['salary_increase_last_year'] < 0.2) * 0.2 +  
    (df['distance_from_home'] > 30) * 0.15 +
    (df['job_position'] == 'Junior') * 0.1 +
    (df['expected_promotion'] == 0) * -0.1 +
    (df['last_evaluation'] < 0.5) * 0.05 +
    (df['number_project'] > 4) * 0.05 +
    (df['time_spend_company'] < 3) * 0.05 +
    (df['Work_accident'] == 1) * 0.05 +
    (df['promotion_last_3_years'] == 0) * 0.05 +
    (df['absent_per_month'] > 5) * 0.05 +
    (df['department'].isin(['DSI', 'technical'])) * 0.05 +
    (df['work_mode'] == 'office') * 0.05 +
    (df['contract_type'] == 'CDD') * 0.05 +
    (df['age'] < 30) * 0.1
)
 
df['left'] = (df['left'] > 0.6).astype(int)
df.head()
```
